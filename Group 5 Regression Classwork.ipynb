```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn import linear_model
```

```python
canada = pd.read_csv ("canada_per_capita_income.csv")
canada
```

Class work predict canada's per capital income in 2020. Using regression modeL and predict the per capita income of canada citizen in year 2020

The dependent variable is the variable to be predicted while the independent variable is the predictor

```python
plt.scatter(canada["year"], canada["per capita income (US$)"])
plt.xlabel("year")
plt.ylabel("per capita income (USD$)")
```

```python
#splitting the columns
#year is the independent variable
year = canada.drop('per capita income (US$)', axis = 1)
year
```

```python
#per capita income (represented by income) is the dependent variable
income = canada.drop("year", axis = 1)
income
```

```python
#create a linear regression object
reg = linear_model.LinearRegression()
reg.fit(year, income)
```

```python
#Using regression modeL and predict the per capita income of canada citizen in year 2020
reg.predict([[2020]])
```

```python
#to get the slope (coefficient) of the regression line.
reg.coef_
```

```python
#to get the intercept of the regression line
reg.intercept_
```

equation for the regression model
y = slope * x + intercept

income = 828.46507522 * year - 1632210.75785546

```python
income = 828.46507522 * 2020 - 1632210.75785546
income
```
