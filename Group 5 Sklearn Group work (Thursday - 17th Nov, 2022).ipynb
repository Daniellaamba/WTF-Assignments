```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#Read in the data
car_prices = pd.read_csv('carprices.csv')
car_prices
```

```python
#Rename the heading 'Car Model' to a readable heading
car_prices.rename(columns = {'Car Model':'Car_Model', 'Sell Price($)':'Sell_Price', 'Age(yrs)':'Age'}, inplace = True)
```

```python
car_prices
```

```python
#Convert Cars to Dummy variables
dummies = pd.get_dummies(car_prices.Car_Model)
dummies
```

```python
#Merge dummy data with initial data
merge = pd.concat([car_prices, dummies], axis = "columns")
merge
```

```python
Drop_data = merge.drop(['Car_Model'], axis = 'columns')
Drop_data
```

```python
final_data = Drop_data.drop(['Audi A5'], axis = 'columns')
final_data
```

```python
#Identify the dependent and independent variables
#Dependent variable = Sell_Price
#Independent Variables = Age, Mileage, Car_Model
```

```python
x = final_data.drop("Sell_Price", axis = "columns")
x
```

```python
y = final_data.Sell_Price
y
```

```python
reg = LinearRegression()
reg
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.predict(x)
```

```python
#1 Predict price of a mercedez benz that is 4 yr old with mileage 45000
reg.predict([[45000,4,0,1]])
```

```python
# Predict price of a BMW X5 that is 7 yr old with mileage 86000
reg.predict([[86000,7,1,0]])
```

```python
#Tell me the score (accuracy) of your model.
reg.score(x,y)
```
